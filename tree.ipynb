```python
import pandas as pd 
import matplotlib.pyplot as plt 

# importing data
data = pd.read_csv('trees.csv',names =['I','Girth','Height','Volume'])


# cleaning data
del data['I']


data.drop(0,inplace=True)
data['Girth']=data['Girth'].astype(float)
data['Height']=data['Height'].astype(float)
data['Volume']=data['Volume'].astype(float)

data['Girth'] *=0.0254       #1 inch = 0.0254 meter
data['Height'] *=0.0305       # 1 foot = 0.305 meters
data['Volume'] *=0.0283      # 1 cubic foot = 0.0283 cub meters


# measuring data
max_girth= (data['Girth'].round(2)).max()
max_height= (data['Height'].round(2)).max()
max_vol= (data['Volume'].round(2)).max()



avg_girth = round(data['Girth'].mean(),2)
avg_height = round( data['Height'].mean(),2)
avg_vol = round(data['Volume'].mean(),2)


med_girth = (data['Girth'].round(2)).median()
med_height = (data['Height'].round(2)).median()
med_vol = (data['Volume'].round(2)).median()

mod_girth = (data['Girth'].round(2)).mode()
mod_height = (data['Height'].round(2)).mode()
mod_vol = (data['Volume'].round(2)).mode()

# tabular form of calculated value

df= pd.DataFrame({'Girth': [max_girth,avg_girth,med_girth,mod_girth[0]],
                  'Height': [max_height,avg_height,med_height,mod_height[0]],
                  'Volume': [max_vol,avg_vol,med_vol,mod_vol[0]]},
                  index = ['Maximum', 'Average', 'Median','Mode'])

# displaying data

print("Tree measurement data \n")
print(data.round(1), end= '\n \n' )
print('Measurements \n')
print(df , end ='\n \n')
print('Graphs')

#visual implementation using matplotlib

plt.plot(data['Girth'])
plt.title('Girth')
plt.ylabel('Meters')
plt.show()

plt.plot(data['Height'])
plt.title('Height')
plt.ylabel('Meters')
plt.show()


plt.plot(data['Volume'])
plt.title('Volume')
plt.ylabel('Cubic Meters')
plt.show()
```
